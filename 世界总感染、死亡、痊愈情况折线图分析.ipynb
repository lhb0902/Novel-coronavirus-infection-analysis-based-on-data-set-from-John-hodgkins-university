```python
#总感染、死亡、痊愈情况折线图分析
#数据包含了一月二十二日到三月十八日的部分
#数据集来源https://github.com/CSSEGISandData/COVID-19
#本次分析是在本地完成的，若要更新数据请注意下方文件路径的替换
#数据集中香港和台湾被单独列出，分析过程中行政区划问题仅是尊重原数据，不代表个人的任何立场
```

```python
#引入所需要的的库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from datetime import datetime
```

```python
#引入文件并打印引入的文件名
root='C:\\Users\\thinkpad\\Desktop\\CSSE COVID-19'
file_names=os.listdir(root)
file_ob_list=[]
Datelist=[]
for file_name in file_names:
    fileob=root+'\\'+file_name
    file_ob_list.append(fileob)
    Datelist.append(file_name[0:10])
print(Datelist)
```

```python
Confirmlist=[]
Recoverlist=[]
Deathlist=[]
print("************************************************")
for file_ob in file_ob_list: 
    df1=pd.read_csv(str(file_ob),keep_default_na=0,dtype=str,index_col='Country/Region')
    df1['Confirmed']=df1['Confirmed'].astype(float)
    sum_Confirmed= df1['Confirmed'].sum()
    print("sum_Confirmed is",sum_Confirmed)
    df1['Deaths']=df1['Deaths'].astype(float)
    sum_Deaths=df1['Deaths'].sum()
    print("sum_Deaths is",sum_Deaths)
    df1['Recovered']=df1['Recovered'].astype(float)
    sum_Recovered=df1['Recovered'].sum()
    print("sum_Recovered is",sum_Recovered)
    
    Confirmlist.append(sum_Confirmed)
    print("以下是历史感染人数")
    print(Confirmlist)
    Deathlist.append(sum_Deaths)
    print("以下是历史死亡人数")
    print(Deathlist)
    Recoverlist.append(sum_Recovered)
    print("以下是历史康复人数")
    print(Recoverlist)
    print("************************************************")
```

```python
#输出总死亡人数散点图
xs = [datetime.strptime(d, '%m-%d-%Y').date() for d in Datelist]
#print(xs)
plt.plot(xs,Deathlist)
plt.xticks(size='small',rotation=90,fontsize=10)
```

```python
#输出总感染人数散点图
plt.plot(xs,Confirmlist)
plt.xticks(size='small',rotation=90,fontsize=10)
```

```python
#输出总康复人数散点图
plt.plot(xs,Recoverlist)
plt.xticks(size='small',rotation=90,fontsize=10)
```
