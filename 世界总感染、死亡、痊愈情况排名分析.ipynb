```python
#世界总感染、死亡、痊愈情况排名分析
#数据包含了一月二十二日到三月十八日的部分
#数据集来源https://github.com/CSSEGISandData/COVID-19
#本次分析是在本地完成的，若要更新数据请注意下方文件路径的替换
#数据集中香港和台湾被单独列出，分析过程中行政区划问题仅是尊重原数据，不代表个人的任何立场
```

```python
#引入所需要的的库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from datetime import datetime
from IPython.display import display_html
```

```python
#引入文件并打印引入的文件名
root='C:\\Users\\thinkpad\\Desktop\\CSSE COVID-19'
file_names=os.listdir(root)
file_ob_list=[]
Datelist=[]
for file_name in file_names:
    fileob=root+'\\'+file_name
    file_ob_list.append(fileob)
    Datelist.append(file_name[0:10])
print(Datelist)
```

```python
#世界日感染人数排名（前五名）
for file_ob in file_ob_list: 
    i=0
    df1=pd.read_csv(str(file_ob),keep_default_na=0,dtype=str,index_col='Country/Region')
    print("*************************",Datelist[i],"感染人数前五名","*************************")
    display_html(df1.sort_values(["Confirmed"],ascending=False).head())
    i=i+1
```

```python
#世界日死亡人数排名（前五名）
for file_ob in file_ob_list: 
    i=0
    df1=pd.read_csv(str(file_ob),keep_default_na=0,dtype=str,index_col='Country/Region')
    print("*************************",Datelist[i],"感染人数前五名","*************************")
    display_html(df1.sort_values(["Deaths"],ascending=False).head())
    i=i+1
```

```python
#世界日痊愈人数排名（前五名）
for file_ob in file_ob_list: 
    i=0
    df1=pd.read_csv(str(file_ob),keep_default_na=0,dtype=str,index_col='Country/Region')
    print("*************************",Datelist[i],"痊愈人数前五名","*************************")
    display_html(df1.sort_values(["Recovered"],ascending=False).head())
    i=i+1
```
